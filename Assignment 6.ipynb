Question 1:You are given two strings s and t. String t is generated by randomly
shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.

```python
def new(s,t):
    for i in s:
        if i not in t:
            return i
        
    return "Both are same"
s = input()
t = input()
new(s,t)
```

Question 2:Given a function that accepts unlimited arguments, check and count how
many data types are in those arguments. Finally return the total in a list.

```python
def unlim(*args):
    lst = []
    count = 0
    count1 = 0
    count2= 0
    count3 = 0
    count4 = 0
    count5 = 0
    for i in args:
        if type(i) == int:
            count = count + 1
        elif type(i) == str:
            count1 = count1 + 1
        elif type(i) == bool:
            count2 = count2 + 1
        elif type(i) == list:
            count3 = count3 + 1
        elif type(i) == tuple:
            count4 = count4 + 1
        elif type(i) == dict:
            count5 = count5 + 1
            
    lst.append(count)
    lst.append(count1)
    lst.append(count2)
    lst.append(count3)
    lst.append(count4)
    lst.append(count5)
    return lst

unlim(1,45,"Hi",False)
```

Qustion 4:Given an integer between 0 and 26, make a variable (self.answer). That
variable would be assigned to a string in this format:
&quot;nines:your answer, threes:your answer, ones:your answer&quot;
You need to find out how many ones, threes, and nines it would at least take
for the number of each to add up to the given integer when multiplied by one,
three or nine (depends).

```python
def ones_threes_nines(n):
    dict1 = {}
    dict1["nines"] = n // 9
    n = n - (dict1["nines"] * 9)
    dict1["threes"] = n // 3
    n = n - (dict1["threes"]*3)
    dict1["ones"] = n
    return dict1
k = int(input())
ones_threes_nines(k)
    
```

think about the implications of running a function recursively. Depending on
the size of n needed and the capabilities of the system in question you might
want to take a different approach.
Write a non-recursive function that takes an integer n and returns the
